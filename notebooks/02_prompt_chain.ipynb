### Prompt Templates-Dynamic Prompts


```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash")
out=StrOutputParser()
```

```python
### My Self Chain Node
def transform_case(result:str):
    return result.upper()
```

```python
prompts=ChatPromptTemplate.from_messages([ 
 {
    "role":"system", "content":"You are a translater and translate input in {language}." }, 
    { "role":"user", "content":"{query}."
     } ])
```

```python
final_prompts = prompts.format_messages(
    language="Hinglish",
    query="Hello, how are you?"
)   
res=llm.invoke(final_prompts)
print(res.content)
```

```python
chains=prompts| llm | out | transform_case
```

```python
res=chains.invoke({"language":"German", "query":"What is your name?"})
print(res)
```
