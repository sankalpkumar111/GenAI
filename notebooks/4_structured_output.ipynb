### Solution is Stuctured Output


```python
from pydantic import BaseModel,Field
from typing import List
class PersonInfo(BaseModel):
    name: str=Field(description="Name of the person")
    city: str=Field(description="City where the person lives")
    age: int=Field(description="Age of the person")
    
structured_llm=llm.with_structured_output(PersonInfo)
```

```python
res1=structured_llm.invoke(f"Pleae give me name ,city and age from the text: {text}")
```

```python
res1.model_dump()
```

```python
class Movie(BaseModel):
    title: str=Field(description="Title of the movie")
    director: str=Field(description="Director of the movie")
    year:int=Field(description="Release year of the movie")

class AllMovies(BaseModel):
    movies: list[Movie]=Field(description="List of movies")

movie_llm=llm.with_structured_output(AllMovies)
```

```python
res=movie_llm.invoke("Give me 3 scifi movies with their director and release year")
```
